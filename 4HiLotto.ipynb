# Chi-Lotto thing
Deciding whether the given data from lotto is probably Gleichverteilt or probably not////////
!!tried!! in Julia

```julia

using CSV, DataFrames, Distributions

# defs und auslesen
df = CSV.read(
        "Lotto.csv",
        DataFrame;
        delim=';', 
        header=false,      
)
rename!(df, Dict(:Column1 => :Zahl, :Column2 => :Häufigkeit)) #Tabellen neu benennen,erste is eig irrelevant
freq = df.Häufigkeit  # data to array
N = sum(freq)  # # aller Ziehungen
m = length(freq)  # Anzahl möglicher Zahlen
erwartet = N / m    # erwartete Häufigkeit pro Zahl bei Gleichverteilung
```

```julia
#Chi Statistik definieren
chisq_stat = sum((freq .- erwartet).^2 ./ erwartet) # Formel
dfree = m - 1   # Freiheitsgrade

#critical and p value
alpha = 0.01
crit = quantile(Chisq(dfree), 1 - alpha) # 99er-Quantil
pval = 1 - cdf(Chisq(dfree), chisq_stat) # p-Wert
```

```julia
#print
println("χ² = ", chisq_stat)
println("kritischer Wert = ", crit)
println("p-Wert = ", pval)

if chisq_stat > crit
    println("Nullhypothese abgelehnt, Gleichverteilt: Probably")
else
    println("Nullhypothese beibehalten, kein direkter Widerspruch lol")
end
```
