# Sampling - Gleichverteilung
Disclaimer: Language is Julia

```julia
using Random, StatsBase, StatsPlots, Plots
```

```julia
samples = rand(0:20, 1000)
mu = mean(samples)
sigma = std(samples)

println(mu , "ist der MW")
println(sigma ,"ist die Standardabweichung") 
```

```julia
N= 20
M = 1000

x= 0:N
count = countmap(samples)

p_emp = get.(Ref(count), x, 0) ./ M
p_theo = fill(1 / (N + 1), length(x))

println(p_theo)
println(p_emp)
```

```julia
bar(x,
     p_emp, 
     label="empirisch", 
     xlabel="Wert", 
     ylabel="p_i")

plot!(x, 
     p_theo, 
     label="theoretisch", 
     lw=2)
```
